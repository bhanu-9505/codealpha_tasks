```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/swiggy_all_menus_india.csv')

# Display first 5 rows
df.head()
```

```python
# Dataset information
df.info()
```

```python
# Statistical summary
df.describe()
```

```python
# Missing values
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))

top_cities = df['City'].value_counts().head(10)

sns.barplot(
    x=top_cities.values,
    y=top_cities.index,
    palette='Oranges_r'
)

plt.title('Top 10 Cities by Restaurant Count')
plt.xlabel('Restaurant Count')
plt.ylabel('City')

plt.show()
```

```python
plt.figure(figsize=(10,6))

top_categories = df['Category'].value_counts().head(10)

sns.barplot(
    x=top_categories.values,
    y=top_categories.index,
    palette='rocket'
)

plt.title('Top Food Categories')
plt.xlabel('Count')
plt.ylabel('Category')

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df['Rating'], bins=20, kde=True)

plt.title('Restaurant Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Price (INR)',
    y='Rating',
    data=df
)

plt.title('Price vs Rating')
plt.xlabel('Price')
plt.ylabel('Rating')

plt.show()
```

# 📌 Key Insights

- Bengaluru has the highest number of restaurants.
- Recommended and Main Course are the most popular food categories.
- Most restaurants have ratings between 4.0 and 5.0.
- Higher-priced restaurants generally tend to receive better ratings.
- Dataset contains no missing values.
